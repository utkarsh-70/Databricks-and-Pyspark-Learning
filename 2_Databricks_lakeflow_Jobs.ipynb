### Databricks Lakeflow Jobs

#### Jobs-Tools for orchestraction 
#### Dag - The order in which everything will be executed.
#### Workflows - Similar to jobs in databricks.
#### Pipelines will be used to process the data along with dag.
#### We can orchestate our pipeline using jobs or databricks workflows.
#### Why databricks jobs -
#### Azure datafactory is used with dabricks or apache airflow for orchestraction and now its alternative is databricks jobs. With databricks jobs now it is equally compatible with all the other clouds as it is happening within databricks.
#### Tasks- Smallest execution which happens is called task.

#### Scenario -1 - Create a job to orchestrate 3 notbooks as task which are linearly dependent on each other.
#### Scenario-2 Create jobs using if else task if condition satisfies run task 1 else run task 2
#### Scenario -3 use  for each task to run the same notebook multiple times-use case i dont know
#### Scenario -4 Dynamic values create 3 task A,B,C A will run first and give some output B will have the parameter and will use return value of A C will run after B and will use return value of A as inline variable.
